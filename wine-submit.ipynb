#### ２）好みの幅について

散布図の確認

```python
x_2 = df["華やか"]
y_2 = df["落ち着き"]

import matplotlib.pyplot as plt

plt.figure(figsize = (5,5))
plt.scatter(x_2, y_2, c = 'y',marker = 'o', s = 50)
plt.grid()
plt.show()
```

エルボー法でクラスタ数を確認

```python
wine_2 = pd.concat([x_2, y_2], axis=1)
SSE_2 = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0)
    km.fit(wine_2)
    SSE_2.append(km.inertia_)

plt.plot(range(1,11), SSE_2, marker='o')
plt.xlabel('k')
plt.ylabel('SSE_2')
plt.show()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3, init = 'random', n_init = 10, max_iter = 300, tol = 1e-04, random_state = 0)

y_km_2 = km.fit_predict(wine_2)

plt.figure(figsize = (5,5))
plt.title('Taste')
plt.xlabel("rich")
plt.ylabel("round")
plt.scatter(wine_2["華やか"], wine_2["落ち着き"], c = y_km_2, marker = 'o', s = 50)
plt.grid()
plt.show()
```
